```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import pysbd
from tqdm import tqdm
import re
import random
import pickle
```

```python
seg = pysbd.Segmenter(language="en", clean=False)

model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

df = pd.read_csv('ParlaMint-GB-lords.csv')
df.head()
```

```python
df['speech'] = df['speech'].astype(str)
df.head()
```

```python
def get_embeddings_df(df):
    speech_embeddings = {}
    
    ids = df['ID'].values
    speeches = df['speech'].values
    
    for i in tqdm(range(len(ids))):
        speech = str(speeches[i])
        
        if len(speech) > 0:
            sentences = seg.segment(speeches[i])

            embeddings = model.encode(sentences)

            speech_embeddings[ids[i]] = embeddings
        
    with open('ParlaMint_GB_lords_embeddings.pkl', 'wb') as f:
        pickle.dump(speech_embeddings, f)
```

```python
get_embeddings_df(df)
```
